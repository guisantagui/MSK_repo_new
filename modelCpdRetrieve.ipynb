```python
mtbH37Rv_mets_ID = []
mtbH37Rv_mets_name = []
for n in range(0, len(mtbH37Rv.metabolites)):
    mtbH37Rv_mets_ID.append(mtbH37Rv.metabolites[n].id)
    mtbH37Rv_mets_name.append(mtbH37Rv.metabolites[n].name)

mtbH37Rv_mets_ID = [i.split('_')[0] for i in mtbH37Rv_mets_ID]
#mtbH37Rv_mets_name = [i.split('_')[0] for i in mtbH37Rv_mets_name]

#print(len(mtbH37Rv_mets_ID))
#print(len(mtbH37Rv_mets_name))

#print(len(list(set(mtbH37Rv_mets_ID))))
#print(len(list(set(mtbH37Rv_mets_name))))
mtbH37Rv_mets = {'seedID': mtbH37Rv_mets_ID, 'name': mtbH37Rv_mets_name}
mtbH37Rv_mets = pd.DataFrame(data=mtbH37Rv_mets)
mtbH37Rv_mets
```

```python
# Import modelSEED dataframe of all the compounds and add a column with just KEGG IDs

import csv
modelSEED = open('C:/Users/Guillem/Documents/PhD/comput/data/ModelSEEDDatabase-master/Biochemistry/compounds.tsv', newline = "")

modelSEED = pd.read_csv(modelSEED, delimiter = "\t")
modelSEED = pd.DataFrame(modelSEED)

modelSEEDKEGGIDs = []
for i in modelSEED['aliases']:
    ID = [x for x in str(i).split('|') if "KEGG" in x]
    if len(ID) == 0:
        modelSEEDKEGGIDs.append('')
    else:    
        modelSEEDKEGGIDs.append([x for x in str(i).split('|') if "KEGG" in x][0].split(': ')[1])

modelSEEDKEGGIDs = pd.Series(modelSEEDKEGGIDs)

modelSEED = pd.concat([modelSEED, modelSEEDKEGGIDs], axis=1)

modelSEED.rename(columns={0:'KEGGIDs'}, 
                 inplace=True)

modelSEED
```

```python
seedName = []
formula = []
mass = []
keggIDs = []
for i in mtbH37Rv_mets['seedID']:
    idx = list(modelSEED['id']).index(i)
    seedName.append(list(modelSEED['name'])[idx])
    formula.append(list(modelSEED['formula'])[idx])
    mass.append(list(modelSEED['mass'])[idx])
    keggIDs.append(list(modelSEED['KEGGIDs'])[idx])
    
mtbH37Rv_mets_extrCols = {'seedName':seedName, 'formula':formula, 'mass':mass, 'KEGGID':keggIDs}
mtbH37Rv_mets_extrCols = pd.DataFrame(data=mtbH37Rv_mets_extrCols)

mtbH37Rv_mets = pd.concat([mtbH37Rv_mets, mtbH37Rv_mets_extrCols], axis=1)

mtbH37Rv_mets = mtbH37Rv_mets.drop(columns=['name'])
mtbH37Rv_mets
```

```python
m2155_mets_ID = []
m2155_mets_name = []
m2155_mets_formula = []
m2155_mets_keggID = []
m2155_mets_seedID = []
for n in range(0, len(m2155.metabolites)):
    m2155_mets_ID.append(m2155.metabolites[n].id)
    m2155_mets_name.append(m2155.metabolites[n].name)
    m2155_mets_formula.append(m2155.metabolites[n].formula)
    if 'KEGG' in m2155.metabolites[n].notes.keys():
        m2155_mets_keggID.append(m2155.metabolites[n].notes['KEGG'][0].split(',')[0])
    else:
        m2155_mets_keggID.append('')
    if 'SEED' in m2155.metabolites[n].notes.keys():
        m2155_mets_seedID.append(m2155.metabolites[n].notes['SEED'][0].split(',')[0])
    else:
        m2155_mets_seedID.append('')    

m2155_mets = {'ID': m2155_mets_ID, 'name': m2155_mets_name, 'formula': m2155_mets_formula, 'keggID': m2155_mets_keggID, 'seedID': m2155_mets_seedID}
m2155_mets = pd.DataFrame(data=m2155_mets)
m2155_mets
```
