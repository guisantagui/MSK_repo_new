Test gene essentiality in both models to see if in the modified model appear new essentialities, that could indicate some vulnerabilities to drugs. 

```python
def compEssent(model1, model2):
    genesRvC = []
    bmFlux = []
    bmFlux_delMod = []
    for g in model1.genes:
        gene = str(g.id)
        with model1:
            model1.genes.get_by_id(gene).knock_out()
            genesRvC.append(gene)
            bmFlux.append(model1.slim_optimize())
        with model2:
            model2.genes.get_by_id(gene).knock_out()
            bmFlux_delMod.append(model2.slim_optimize())
    modelEssntlty = pd.DataFrame(list(zip(genesRvC, bmFlux, bmFlux_delMod)), columns = ['gene', 'biomass_flux_intact', 'biomass_flux_mut']) 
    return modelEssntlty
```

```python
iEK1011inVivoEssntlty = compEssent(iEK1011inVivo, iEK1011inVivo_noGenes)
iEK1011inVivoEssntlty
```

```python
# Filter the dataframe to keep genes with altered biomass fluxes
iEK1011inVivoEssntlty = iEK1011inVivoEssntlty[iEK1011inVivoEssntlty[['biomass_flux_intact', 'biomass_flux_mut']].nunique(axis = 1).gt(1)]
iEK1011inVivoEssntlty
```

Check if the essentiality of these genes depend on a single compound in the medium by removing it one by one 

```python
inVivoMedCache = iEK1011inVivo.medium
```

```python
remCpd = []
NoDelsModBmFlxs = []
Rv1380DelBmFlxs = []
Rv0382cDelBmFlxs = []
iEK1011inVivo.medium = inVivoMedCache
iEK1011inVivoRv1380Del.medium = inVivoMedCache
iEK1011inVivoRv0382cDel.medium = inVivoMedCache
for cpd in iEK1011inVivoRv0382cDel.medium:
    inVivoMedium = copy.copy(inVivoMedCache)
    remCpd.append(cpd)
    inVivoMedium[cpd] = 0.0
    iEK1011inVivo.medium = inVivoMedium
    iEK1011inVivoRv1380Del.medium = inVivoMedium
    iEK1011inVivoRv0382cDel.medium = inVivoMedium
    bmFlxNoDelsMod = iEK1011inVivo.slim_optimize()
    bmFlxRv1380Del = iEK1011inVivoRv1380Del.slim_optimize()
    bmFlxRv0382cDel = iEK1011inVivoRv0382cDel.slim_optimize()
    print("Biomass Flux of NoDelsMod without " + cpd + ": " + str(bmFlxNoDelsMod))
    print("Biomass Flux of Rv1380Del without " + cpd + ": " + str(bmFlxRv1380Del))
    print("Biomass Flux of Rv0382cDel without " + cpd + ": " + str(bmFlxRv0382cDel))
    iEK1011inVivo.medium = inVivoMedCache
    iEK1011inVivoRv1380Del.medium = inVivoMedCache
    iEK1011inVivoRv0382cDel.medium = inVivoMedCache
    NoDelsModBmFlxs.append(bmFlxNoDelsMod)
    Rv1380DelBmFlxs.append(bmFlxRv1380Del)
    Rv0382cDelBmFlxs.append(bmFlxRv0382cDel)
compMedRem = pd.DataFrame(list(zip(remCpd, NoDelsModBmFlxs, Rv1380DelBmFlxs, Rv0382cDelBmFlxs)), columns = ['cpd removed from media', 'NoDelsModBmFlxs', 'Rv1380DelBmFlx', 'Rv0382cDelBmFlx'])

compMedRem
```

The genes Rv1380 and Rv0382c become essential when alanine, asparagine and aspartate are removed from in vivo medium

```python
# Let's try if there are combinations of 2 AAs that make Rv1380 and Rv0382c essential when removed from the medium
n = 1
AAPairLst = []
for i in range(0, len(inVivoAAs)):
    for j in range(n, len(inVivoAAs)):
        pair = [inVivoAAs[i], inVivoAAs[j]]
        AAPairLst.append(pair)
    n +=1
```

```python
iEK1011inVivoRv1380Del.medium = inVivoMedCache
iEK1011inVivoRv0382cDel.medium = inVivoMedCache
inVivoMedium = copy.copy(inVivoMedCache)
for aaP in AAPairLst:
    for aa in aaP:
        inVivoMedium[aa] = 0.0
    iEK1011inVivo.medium = inVivoMedium
    iEK1011inVivoRv1380Del.medium = inVivoMedium
    iEK1011inVivoRv0382cDel.medium = inVivoMedium
    print(str(aaP) + " fullModel: " + str(iEK1011inVivo.slim_optimize())) 
    print(str(aaP) + " Rv1380Del: " + str(iEK1011inVivoRv1380Del.slim_optimize()))
    print(str(aaP) + " Rv1380Del: " + str(iEK1011inVivoRv0382cDel.slim_optimize()))
    inVivoMedium = copy.copy(inVivoMedCache)
    iEK1011inVivo.medium = inVivoMedCache
    iEK1011inVivoRv1380Del.medium = inVivoMedCache
    iEK1011inVivoRv0382cDel.medium = inVivoMedCache
```
