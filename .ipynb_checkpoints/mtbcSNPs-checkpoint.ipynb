```python
annotGrntx = allel.FeatureTable.from_gff3('C:/Users/Guillem/Documents/PhD/comput/data/MTBC_ancestralAnnot/genes.gff', attributes = ["Name","locus_tag", "product"])
```

```python
annotGrntx
```

```python
def retrPos(gff, geneLst):
    if type(geneLst) == str:
        geneLst = [geneLst]
    startLst = []
    endLst = []
    for i in range(0, len(gff['locus_tag'])):
        for g in geneLst:
            if(gff['locus_tag'][i] == g):
                startLst.append(gff['start'][i])
                endLst.append(gff['end'][i])
    genePos = pd.DataFrame(list(zip(geneLst, startLst, endLst)), columns = ['gene', 'start', 'end'])
    return genePos
```

```python
plcABCPos = retrPos(annotGrntx, ['Rv2349c', 'Rv2351c', 'Rv2350c'])
```

```python
ann = []
gene = []
ntSubs = []
aaSubs = []
for i in range(0, len(hetSNPs['variants/ANN'])):
    ann.append(hetSNPs['variants/ANN'][i].split('|')[1])
    gene.append(hetSNPs['variants/ANN'][i].split('|')[3].split('_')[0])
    ntSubs.append(hetSNPs['variants/ANN'][i].split('|')[9])
    aaSubs.append(hetSNPs['variants/ANN'][i].split('|')[10])
```

```python
hetSNPsDF = pd.DataFrame(list(zip(hetSNPs['variants/REF'], hetSNPs['variants/ALT'], hetSNPs['variants/POS'], ann, gene, ntSubs, aaSubs)), columns = ['ref', 'alt', 'pos', 'ann', 'gene', 'ntSubs', 'aaSubs'])
```
