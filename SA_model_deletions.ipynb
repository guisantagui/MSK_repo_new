Do copies of the models and delete all of the involved reactions to see how biomass flux is afected. 

```python
import copy

iSA863_nadQdel = copy.deepcopy(iSA863)
iSB619_nadQdel = copy.deepcopy(iSB619)
iYS854_nadFadQdel = copy.deepcopy(iYS854)
```

```python
# When removed both biomass flux approximates 0
iSA863_nadQdel.optimize()
```
